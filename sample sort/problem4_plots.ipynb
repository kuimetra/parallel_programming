```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_seq = np.genfromtxt('log_seq.csv', delimiter=',', dtype=float, filling_values=0)[8:16, :]
data_par = np.genfromtxt('log_par.csv', delimiter=',', dtype=float, filling_values=0)[8:16, :]
data_speed = np.genfromtxt('log_speedup.csv', delimiter=',', dtype=float, filling_values=0)[8:16, :]

data_weak_scaling = np.genfromtxt('log_weak_scaling.csv', delimiter=',', dtype=float, filling_values=0)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 10))

ax[0][0].plot(data_seq[:, 1], data_seq[:, 2], label="seq")
ax[0][1].plot(data_seq[:, 1], data_seq[:, 3], label="seq")
ax[1][0].plot(data_seq[:, 1], data_seq[:, 4], label="seq")
ax[1][1].plot(data_seq[:, 1], data_seq[:, 5], label="seq")

ax[0][0].plot(data_par[:, 1], data_par[:, 2], label="par")
ax[0][1].plot(data_par[:, 1], data_par[:, 3], label="par")
ax[1][0].plot(data_par[:, 1], data_par[:, 4], label="par")
ax[1][1].plot(data_par[:, 1], data_par[:, 5], label="par")

ax[0][0].plot(data_speed[:, 1], data_speed[:, 2], label="speedup")
ax[0][1].plot(data_speed[:, 1], data_speed[:, 3], label="speedup")
ax[1][0].plot(data_speed[:, 1], data_speed[:, 4], label="speedup")
ax[1][1].plot(data_speed[:, 1], data_speed[:, 5], label="speedup")

ax[0][0].set_title('Part 1')
ax[0][0].set_xlabel('p')
ax[0][1].set_title('Part 2')
ax[0][1].set_xlabel('p')
ax[1][0].set_title('Part 3')
ax[1][0].set_xlabel('p')
ax[1][1].set_title('Part 4')
ax[1][1].set_xlabel('p')

ax[0][0].get_yaxis().set_visible(False)
ax[0][1].get_yaxis().set_visible(False)
ax[1][0].get_yaxis().set_visible(False)
ax[1][1].get_yaxis().set_visible(False)

fig.suptitle("Execution time of each part for n = 50.000.000", fontsize=18)

plt.legend()
plt.show()
```

```python
plt.plot(data_seq[:, 1], data_seq[:, 6], label="seq")
plt.plot(data_par[:, 1], data_par[:, 6], label="par")
plt.plot(data_speed[:, 1], data_speed[:, 6], label="speedup")

plt.title("Sample sort execution time for n = 50.000.000")
plt.xlabel('p')
plt.yticks([])
plt.legend()
plt.show()
```

```python
plt.plot(data_weak_scaling[:, 1], data_weak_scaling[:, 6])

plt.title("Weak scaling")
plt.xlabel('p')
plt.yticks(data_weak_scaling[:, 6])
plt.show()
```
